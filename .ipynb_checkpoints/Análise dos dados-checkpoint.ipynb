```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Função para formatar a matriz de correlação
def plot_corr_matrix(corr_matrix, size=(10, 10)):
  # Set up the matplotlib figure
  fig, ax = plt.subplots(figsize=size)

  # Generate a custom diverging colormap
  cmap = sns.diverging_palette(220, 10, as_cmap=True)

  # Generate a mask for the upper triangle
  mask = np.triu(np.ones_like(corr_matrix.corr(), dtype=np.bool))

  # Draw the heatmap with the mask and correct aspect ratio
  sns.heatmap(corr_matrix.corr(), mask=mask, cmap=cmap, vmax=.99, center=0,
              square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot = True)
```

```python
data = pd.read_json('Dados\BaseLimpa.json')
```

```python
data.head(2)
```

```python
simnao = {
    'Sim' : 1,
    'Não' : 0
}
data['Churn'].replace(simnao, inplace = True)
```

```python
cols = data.columns.tolist()
cols.remove('Churn')
cols.append('Churn')
cols
```

```python
matrix = data[cols].replace(simnao).corr()
plot_corr_matrix(matrix)
```

```python
# Variavel de interesse
target = 'Churn'
print('target:', target)

# Variaveis numericas
num_vars = data.select_dtypes(include=['int', 'float']).columns.tolist()
num_vars.remove(target)
num_vars.remove('Idoso')
print(f'numerical vars({len(num_vars)}): {num_vars}')

# Variaveis categorical
cat_vars = data.select_dtypes(exclude=['int', 'float']).columns.tolist()
cat_vars.append('Idoso')
print(f'categorical vars({len(cat_vars)}): {cat_vars}')
```

```python
# Configurando a figura do matplot lib
fig = plt.figure(figsize=(25,90))
fig.subplots_adjust(hspace=0.8, wspace=0.4)

total_plots = len(cat_vars)
i = 0

custom_params = {"axes.spines.right": False, "axes.spines.top": False,
                     "axes.spines.left": False, "axes.spines.bottom": False}

sns.set_theme(style="ticks", rc=custom_params)

# Iterando para todas variaveis categoricas
for var in cat_vars:
    # Pegando o index certo para plotar o grafico
    ax = fig.add_subplot(total_plots, 2, i+1)
    # Desenhando o grafico do tipo violinplot
    
    sns.countplot(x=var, hue=target, palette='turbo',data=data, ax=ax)
    ax.set_title(var)
    ax.set_ylabel('')
    ax.set_xlabel('')
    i += 1
    for container in ax.containers:
        ax.bar_label(container, fontsize=15, padding=-17, color='white')

plt.show()
```

```python
for col in num_vars:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=target, y=col, data=data)
    plt.title(f"Boxplot de {col} por Churn", fontsize=16)
    plt.ylabel(col, fontsize=14)
    plt.xlabel('Churn', fontsize=14)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.show()
```

## Resumo
<ul>
<li> Na correlação observamos um alto impacto das variáveis Parceiro(alterado para numérica), Fidelidade e Despesa total em Churn de maneira negativa.</li>
<br>
<li> Nas comparações de variáveis categoricas, podemos citar que caso o cliente não tenha parceiro ele tende a cometer churn numa taxa de um terço, já para aqueles que não tem dependentes o churn tem uma chance de 25%. Vemos uma alta taxa de churn na fibra ótica, falta de segurança online, falta de suporte técnico, falta de proteção no dispositivo e falta de backup. Mas pode-se dizer que o contrato mensal e a fatura por débito eletrõnico é os principais em porcentagem de churn (próximo aos 50%).</li>
<br>
<li> Nos boxplot vemos que a alta de gasto mensal gera uma maior probabilidade de churn, de mesma forma o gasto diário. Já a despesa total é o contrário, quanto maior o gasto total, menos se espera um churn. Também atrelado a despesa total temos a fidelidade, que nada mais é que o tempo em que o cliente está linkado a empresa.</li>
</ul>
