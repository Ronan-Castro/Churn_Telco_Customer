```python
data = pd.read_json('Dados\Telco-Customer-Churn.json')
data
```

```python
customer = pd.json_normalize(data.customer)
internet = pd.json_normalize(data.internet)
phone = pd.json_normalize(data.phone)
account = pd.json_normalize(data.account)
```

```python
customer['Churn'] = data.Churn
```

```python
data1 = pd.concat([customer, internet], axis=1)
data2 = pd.concat([phone, account], axis=1)
data_final = pd.concat([data1, data2], axis=1)
```
